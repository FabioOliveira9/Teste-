🤔 **Mas como saber se a $f_X(x)$ escolhida é adequada?**




### Teste de hipótese:

Em geral, o processo para determinar se a $f_x(x)$ escolhida é adequada aos dados **observados**, é dado pelas seguintes etapas:

* Coletar o histograma da amostra com $n$ observações, e plotar o histograma de densidade.

* Escolher uma função de densidade de probabilidade, chamada de **modelo**, dentre uma coleção de famílias de distribuições conhecidas.

* Realizar o ajuste da curva do modelo com o histograma de densidade de probabilidade ([Manual](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html)).

* Para um histograma com $k$ bins, $p_{o,i}$ é a **probabilidade observada** associada ao `bin` $i$ do histograma onde $i=1,2,...,k$. Por outro lado, $p_{m,i}$ é a **probabilidade esperada** fixada pelo modelo probabilistico no centro do bin $i$ do histograma.

* Se a hipótese de que o modelo é $f_X(x)$ é **verdadeira**, então:
$$p_{o,1}=p_{m,1}, p_{o,2}=p_{m,2},...,p_{o,k}=p_{m,k}$$

* Por outro lado, se existe pelo menos uma diferença, ou um desvio suficientemente considerável, então a hipótese de que o modelo é uma $f_X(x)$ é **falsa**.


Para verificar se nossa hipótese é verdadeira, precisamos formular duas hopóteses excludentes, denotadas por:

* **Hipótese nula ($\mathcal{H}_0$):** Afirmação sobre a distribuição ou o parâmetro da população que será testada. \\
🚀 **Exemplo:** A amostra segue uma distribuição gaussiana ou as probabilidades observadas são iguais as probabilidades esperadas.
* **Hipótese alternativa ($\mathcal{H}_1$):** Afirmação oposta à hioítese nula, que será aceita se a hipótese nula for rejeitada. \\
🚀**Exemplo:** A amostra não segue uma distribuição gaussiana

Para testar ambas as hipóteses, precisamos determinar o **teste estatístico** (por exemplo, o teste do $\chi^2$).


Além disso, precisamos, também, definir o erro esperado para esse teste de hipótese. Podemos definir dois tipos de erros na probabilidade de cometer um erro na **tomada de decisão**:

* **Probabilidade de erro de tipo I:** é definido como:
$$\alpha = P(\text{erro tipo I}) = P(\text{rejeitar $\mathcal{H}_0$}|\text{$\mathcal{H}_0$ é verdadeira})$$
onde $\alpha$ é chamado de **nível de confiança** ou tamanho do teste.

* **Probabilidade de erro do tipo II** é definido como:
$$\beta = P(\text{erro tipo II}) = P(\text{não rejeitar $\mathcal{H}_0$}|\text{$\mathcal{H}_0$ é falsa})$$

Na prática, o $\alpha$ é a porcentagem da distribuição de $\mathcal{H}_0$ que estamos admitindo errar. Ou seja, para um $\alpha=0,05$, se o valor do **teste estatístico** "cair dentro" de 95% da distribuição de $\mathcal{H}_0$, então devemos **não rejeitar a hipótese nula**. Por outro lado, se o valor do teste estatístico cair na região de erro de tipo I, ou fora dos limites de 95% da distribuição, então devemos **rejeitar a hipótese nula**.



 **🔥Atenção🔥**:  Apenas o valor de $\alpha$ é usado nos testes estatísticos onde tipicamente é $0,05$ (ou 5%).



### Teste do $\chi^2$:

Um dos **testes estatísticos** mais conhecidos para avaliar se uma distribuição (ou modelo) está aderente a amostra (ou o histograma de densidade) é o **teste de $\chi^2$**, também chamado de **teste de aderência**.

Seja um histograma com $k$ bins, o $\chi_o^2$ observado na amostra pode ser definido como:

$$\chi_o^2 = \sum_{i=1}^k \frac{(O_I - E_i)^2}{E_i}$$

onde $i=1,2,...,k$ bins do histograma, $E_i = p_{m,i}\cdot n$ é o número de ocorrências esperadas para o bin $i$, $O_i = p_{o,i}\cdot n$ é o número de observações para o bin $i$ e $n$ é o número total de observações da amostra.

Em um histograma com $k$ bins, a probabilidade observada ($p_{o,i}$) do bin $i$ é dada pela área da barra vertical do histograma. Ou seja, é o produto entre a dendidade de probabilidade do bin $i$ (altura) e o tamanho do bin $i$ (base).

Da mesma forma, a probabilidade esperada ($p_{m,i}$) do bin $i$ é dada pela área abaixo da curva do modelo testado entre o valor superior e inferior (ou o intervalo) do bin $i$ do histograma. Na prática, a probabilidade esperada é o produto entre  o valor fixado pelo modelo no centro do bin $i$ (altura) e o tamanho do bin $i$ (base).

### 1) Cálculo de densidade:

O histograma é um gráfico de barras onde a altura de uma barra representa a contagem de observações em um determinado intervalo. Por sua vez, o eixo horizontal é formado por conjuntos $k $de intervalos, chamados de `bins`.

A densidade de cada bin do histograma é definida como:

$$d(b_i) = \frac{\text{count($i$)}}{\left(\sum_{i=1}^k \text{count($b_i$)}\right)\cdot \text{width($b_i$)} }$$

onde $i$ é o índice do bin, $k$ é o número total de bins do histograma, count($b_i$) é uma função que retorna o número de observações do bin i e width($b_i$) é uma função que retorna o tamanho do bin $i$ no eixo horizontal.

Crie uma função para calcular a densidade de cada `bin` de um histograma utilizando o protótipo da função abaixo.

**🔥Atenção🔥:** A flag `density` da função `np.histogram` deve permancer `False`.


```python
# Igual ao experimento 1
def get_density( data, nbins=50, xmin=-6, xmax=6):
  freq_obs, bin_edges = np.histogram(data, bins=nbins, range=(xmin,xmax), density=False)
  tot_obs             = np.sum(freq_obs)
  bin_sizes           = np.diff(bin_edges)
  density_obs         = freq_obs / (tot_obs * bin_sizes)
  return density_obs, freq_obs, bin_edges
```

```python
def get_chi2( X , model, nbins=50 , xmin=-6, xmax=6):

  n                  = np.sum(get_density(X, nbins=nbins, xmin=xmin, xmax=xmax)[1])
  degrees_of_freedom = nbins - 1
  density_obs, freq_obs, bin_edges = get_density(X, nbins=nbins, xmin=xmin, xmax=xmax)
  bin_centers      = (bin_edges[:-1] + bin_edges[1:]) / 2
  bin_sizes        = np.diff(bin_edges)
  prob_obs         = density_obs * bin_sizes


  params, _          = curve_fit(model, bin_centers, density_obs )
  x                  = np.linspace(bin_edges[0], bin_edges[-1], 10000)
  density_expected   = model(bin_centers, *params)

  density_expected_for_each_bin_center = model(bin_centers, *params)
  prob_expected      = density_expected * bin_sizes
  freq_expected      = prob_expected * n

  chi2 = 0
  for f_obs, f_exp in zip(freq_obs, freq_expected):
    if f_exp > 0: # só vai computar se a frequência esperada for positiva para evitar divisão por zero
      chi2 += (f_obs - f_exp) ** 2 / f_exp


  output = {'hist':{
                  'density_obs': density_obs,
                  'bin_edges'  : bin_edges,
                  'bin_centers': bin_centers,
                  'bin_sizes'  : bin_sizes,
                  'prob_obs'   : prob_obs,
                  'freq_obs'   : freq_obs,
                  },
            'fit':{
                  'x'               : x,
                  'density_expected': density_expected,
                  'params'          : params,
                  'density_expected_for_each_bin_center': density_expected_for_each_bin_center,
                  'prob_expected'   : prob_expected,
                  'freq_expected'   : freq_expected,
                  }
            }

  return chi2, degrees_of_freedom, output
```
